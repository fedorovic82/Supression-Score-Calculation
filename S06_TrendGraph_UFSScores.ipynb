```
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```
is_countrySpecific = 0
countrycode        = {"FR","US"}
timesteps          = 5
destination_table  = "data_user_menaka.tracking_ufs_suppression_scores_overtime"
```

```
ufs_scores       = spark.sql("""show tables in data_suppression""")
ufs_scores       = ufs_scores.where(col("tableName").rlike("perfect_golden_ufs_recipientscores"))
ufs_scores       = ufs_scores.withColumn("date", substring(col("tableName"),36,100))

window           = Window.partitionBy('database').orderBy(col('date').desc())
ufs_scores       = ufs_scores.withColumn("rank",rank().over(window))
ufs_scores       = ufs_scores.where(col("rank") <= timesteps)
display(ufs_scores)
```

```
rank1 = "data_suppression." + ufs_scores.where(col("rank") == 1).select("tableName").rdd.max().tableName
current = spark.table(rank1)
current = current.withColumn("CreatedDate", lit(rank1))
current = current.withColumn("CreatedDate", substring(col("CreatedDate"),53,100))
current = current.withColumn("CreatedDate", from_unixtime(unix_timestamp(substring(col("CreatedDate"),0,8), 'yyyyMMdd'),'yyyy-MM-dd'))
#current = current.withColumn("Score", when( (col("Score") == 7) & (col("Opens_Less_6_Months") > 0), '7_With_Opens').otherwise(col("Score")))

if is_countrySpecific:
  current = current.where(col("cp_countryCode").isin(countrycode))


timestep_0_overview = current.groupby("CreatedDate","cp_countryCode","Score").agg(countDistinct(col("cp_ohubId")).alias("NoofRecipients")).sort("cp_countryCode","Score")
```

```
rank2 = "data_suppression." + ufs_scores.where(col("rank") == 2).select("tableName").rdd.max().tableName
current_minus_one = spark.table(rank2)
current_minus_one = current_minus_one.withColumn("CreatedDate", lit(rank2))
current_minus_one = current_minus_one.withColumn("CreatedDate", substring(col("CreatedDate"),53,100))
current_minus_one = current_minus_one.withColumn("CreatedDate", from_unixtime(unix_timestamp(substring(col("CreatedDate"),0,8), 'yyyyMMdd'),'yyyy-MM-dd'))
#current_minus_one = current_minus_one.withColumn("Score", when( (col("Score") == 7) & (col("Opens_Less_6_Months") > 0), '7_With_Opens').otherwise(col("Score")))

if is_countrySpecific:
  current_minus_one = current_minus_one.where(col("cp_countryCode").isin(countrycode))
  
timestep_minus_one_overview = current_minus_one.groupby("CreatedDate","cp_countryCode","Score").agg(countDistinct(col("cp_ohubId")).alias("NoofRecipients")).sort("cp_countryCode","Score")
```

```
rank3 = "data_suppression." + ufs_scores.where(col("rank") == 3).select("tableName").rdd.max().tableName
current_minus_two = spark.table(rank3)
current_minus_two = current_minus_two.withColumn("CreatedDate", lit(rank3))
current_minus_two = current_minus_two.withColumn("CreatedDate", substring(col("CreatedDate"),53,100))
current_minus_two = current_minus_two.withColumn("CreatedDate", from_unixtime(unix_timestamp(substring(col("CreatedDate"),0,8), 'yyyyMMdd'),'yyyy-MM-dd'))
#current_minus_two = current_minus_two.withColumn("Score", when( (col("Score") == 7) & (col("Opens_Less_6_Months") > 0), '7_With_Opens').otherwise(col("Score")))

if is_countrySpecific:
  current_minus_two = current_minus_two.where(col("cp_countryCode").isin(countrycode))
  
timestep_minus_two_overview = current_minus_two.groupby("CreatedDate","cp_countryCode","Score").agg(countDistinct(col("cp_ohubId")).alias("NoofRecipients")).sort("cp_countryCode","Score")
```

```
rank4               = "data_suppression." + ufs_scores.where(col("rank") == 4).select("tableName").rdd.max().tableName
current_minus_three = spark.table(rank4)
current_minus_three = current_minus_three.withColumn("CreatedDate", lit(rank4))
current_minus_three = current_minus_three.withColumn("CreatedDate", substring(col("CreatedDate"),53,100))
current_minus_three = current_minus_three.withColumn("CreatedDate", from_unixtime(unix_timestamp(substring(col("CreatedDate"),0,8), 'yyyyMMdd'),'yyyy-MM-dd'))
#current_minus_three = current_minus_three.withColumn("Score", when( (col("Score") == 7) & (col("Opens_Less_6_Months") > 0), '7_With_Opens').otherwise(col("Score")))

if is_countrySpecific:
  current_minus_three = current_minus_three.where(col("cp_countryCode").isin(countrycode))
  
timestep_minus_three_overview = current_minus_three.groupby("CreatedDate","cp_countryCode","Score").agg(countDistinct(col("cp_ohubId")).alias("NoofRecipients")).sort("cp_countryCode","Score")
```

```
rank5               = "data_suppression." + ufs_scores.where(col("rank") == 5).select("tableName").rdd.max().tableName
current_minus_four = spark.table(rank5)
current_minus_four = current_minus_four.withColumn("CreatedDate", lit(rank5))
current_minus_four = current_minus_four.withColumn("CreatedDate", substring(col("CreatedDate"),53,100))
current_minus_four = current_minus_four.withColumn("CreatedDate", from_unixtime(unix_timestamp(substring(col("CreatedDate"),0,8), 'yyyyMMdd'),'yyyy-MM-dd'))
#current_minus_four = current_minus_four.withColumn("Score", when( (col("Score") == 7) & (col("Opens_Less_6_Months") > 0), '7_With_Opens').otherwise(col("Score")))


if is_countrySpecific:
  current_minus_four = current_minus_four.where(col("cp_countryCode").isin(countrycode))
  
timestep_minus_four_overview = current_minus_four.groupby("CreatedDate","cp_countryCode","Score").agg(countDistinct(col("cp_ohubId")).alias("NoofRecipients")).sort("cp_countryCode","Score")
```

```
combined = timestep_0_overview.union(timestep_minus_one_overview).union(timestep_minus_two_overview).union(timestep_minus_three_overview).union(timestep_minus_four_overview)
combined = combined.sort("CreatedDate", ascending = False)
```

```
reshaped_df = combined.groupby('cp_countryCode','Score').pivot('CreatedDate').agg(max(col("NoofRecipients")).alias("NoofRcp"))
```

```
reshaped_df.write.mode("overwrite").saveAsTable(destination_table)
```
